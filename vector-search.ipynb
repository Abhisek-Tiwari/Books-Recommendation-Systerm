```python
from langchain.embeddings import HuggingFaceEmbeddings
from langchain_chroma import Chroma
from langchain_text_splitters import CharacterTextSplitter
from langchain_community.document_loaders import TextLoader
```

```python
from dotenv import load_dotenv
load_dotenv()
```

```python
import pandas as pd

books = pd.read_csv('book_cleaned_no_double_quotes.csv')
```

```python
books
```

```python
books['tagged_description'].head(4)
```

```python
books['tagged_description'].to_csv('tagged_description.txt',
                                   sep = '\n',
                                   index = False,
                                   header = False)
```

```python
raw_documents = TextLoader('tagged_description.txt', encoding='utf-8').load()
text_splitter = CharacterTextSplitter(chunk_size=0, chunk_overlap=0, separator='\n')
documents = text_splitter.split_documents(raw_documents)
```

```python
documents[0]
```

```python
# Load an open-source embedding model
embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
db_books = Chroma.from_documents(
    documents,
    embedding=embedding_model
)
```

```python
query = 'A book about cats'
docs = db_books.similarity_search(query, k=7)
```

```python
docs
```

```python
books[books['isbn13'] == int(docs[0].page_content.split()[0].strip())]
```

```python
# Make a function to implement this
def retrieve_semantic_recommendation(
        query: str,
        k: int = 10,
)-> pd.DataFrame:
    recs = db_books.similarity_search(query, k= 50)

    books_list = []

    for i in range(0,len(recs)):
        books_list += [int(recs[i].page_content.strip('"').split()[0])]

    return books[books['isbn13'].isin(books_list)].head(k)
```

```python
retrieve_semantic_recommendation('A book about cats', 8)
```
